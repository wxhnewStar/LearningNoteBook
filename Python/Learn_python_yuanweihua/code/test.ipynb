```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

iris = sns.load_dataset("iris")
```

```python
print(type(iris))

print(iris.head())
```

```python
import pandas as pd
iris[["sepal_length","sepal_width"]].plot.hist()
```

```python
iris.plot.hist(cmap=plt.cm.Accent, alpha=0.5, bins=20, stacked=False)
```

```python
iris['petal_length'].plot.line()
```

```python
print(iris["species"].astype("category").value_counts())
print(type(iris["species"].astype("category").cat.codes))
print(type(iris["species"].astype("category")))

# astype 转换后的类型也是 series
# series 的单个值的类型是 category,对 catetegory 可以访问其 cat 属性
# cat 属性的 codes 属性可以访问这个 series 的所有值的 code(整数代码)
# cat 属性的 categories 属性可以访问这个 series 的所有值原始类别值
```

```python
# 将 species 转换成数字
changed_iris = iris.copy()
changed_iris['species'] = changed_iris['species'].astype('category').cat.codes
print(changed_iris.head())

changed_iris.plot.scatter(x='sepal_length', y='sepal_width', c='species', cmap=plt.cm.Accent)
```

```python
iris['species'].value_counts().plot.pie(legend=True)
```
