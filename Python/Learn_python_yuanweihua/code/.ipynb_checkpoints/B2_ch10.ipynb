```python
import matplotlib.pyplot as plt
import numpy as np

x_array = np.linspace(0, 2*np.pi, 100)

print(x_array)

print(f"type of x_array: {type(x_array)}") 
"""
class: numpy::ndarray
n-dimensional array
attribute: shape, dtype, ndim, size, itemsize, nbytes
"""
# print((x_array.shape, x_array.dtype, x_array.ndim, x_array.size, x_array.itemsize, x_array.nbytes))
print(np.rad2deg(x_array))


sin_y = np.sin(x_array)

cos_y = np.cos(x_array)

# 设置图片大小
fig, ax = plt.subplots(figsize=(8,6))
"""
fig: figure object
ax: axes object (axes object is a part of figure object)
通过 ax 轴对象,可以对图像进行设置
"""


# 绘制正弦和余弦曲线
ax.plot(x_array, sin_y,label ='sin', color='b', linewidth=2)
ax.plot(x_array, cos_y,label ='cos', color='r', linewidth=2)

# 设置标题 横轴和纵轴标签
ax.set_title('sine and cosine functions')
ax.set_xlabel('x')
ax.set_ylabel('y')

# 添加图例
ax.legend()

# 设置横轴和纵轴范围
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)

# 设置横轴刻度
x_ticks = np.arange(0, 2*np.pi + np.pi/2, np.pi/2)
x_ticklabels = [r'$0$', r'$\frac{\pi}{2}$', r'$\pi$',
                r'$\frac{3\pi}{2}$', r'$2\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)

# 横纵轴采用相同的 scale
ax.set_aspect('equal')
plt.grid()

# 将图片存为 SVG 格式
plt.savefig('sine_cosine.svg', format='svg')

# 显示图像
plt.show()

```
