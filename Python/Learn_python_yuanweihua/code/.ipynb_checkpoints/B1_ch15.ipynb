## 广播原则

简单来说，NumPy 的广播原则 (broadcasting) 指定了不同形状的数组之间的算术运算规则，将形状较小的数组扩展为与形状较大的数组相同，再进行运算，以提高效率。

### 一维数组和标量
等于拓展标量到一维数组然后相加

### 一维数组和列向量
得到的结果是一个二维向量,等价于先把一维数组和列向量拓展到二维再相加

### 二维数组和标量
相当于把标量拓展到二维数组,再对应位置相加

### 二维数组和一维数组
相当于先把一维数组,向下复制拓展成二维数组,再逐位相加

### 二维数组和列向量
相当于把列向量,向右复制拓展成形状一样的二维数组,再逐位相加

```python
a_1D ** 2
```

```python
a_1D % 2
```

## 数学函数

常量有:
* np.pi # 不是函数
* np.e
* np.inf # 正无穷
* np.NAN # 非数

> numpy.NAN（或np.nan）是 NumPy 库中表示缺失值（NaN，Not a Number）的特殊值。它通常用于以下情况：
> * 缺失值标记：numpy.NAN 可以用作数组中缺失值的标记。当数组中某个元素的值未知、无效或不可用时，可以使用 numpy.NAN 来表示该缺失值。
> * 数值计算：numpy.NAN 在数值计算中具有传播性质。即，任何与 numpy.NAN 进行数学运算的结果都将是 numpy.NAN。这对于处理包含缺失值的数据并进行数学运算或统计计算非常有用。
> * 数据清洗和处理：在数据处理过程中，通常需要处理包含缺失值的数据。numpy.NAN 可以用于标记和处理缺失值，例如进行缺失值填充、删除或计算缺失值的统计量等。
> * 逻辑判断：numpy.NAN 在逻辑判断中具有特殊的性质。与 numpy.NAN 进行任何逻辑比较操作（如等于、不等于、大于、小于等），结果都将是 False。这是因为缺失值的具体数值未知，无法进行确定的逻辑比较。

其他函数:
* 三角函数就不说了
* np.power(x, 2)
* np.exp(x) # e^x
* np.log(x) # logx
* np.abs(x)  # 绝对值函数
* np.floor(x) # 向下取整数
* np.ceil(x) # 向上取整数
* np.sgn(x) # 大于 0 的值都是 1,小于 0 的都是 -1

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys
# 自定义可视化函数
def visualize_fx(x_array, f_array, title, step = False):
    fig, ax = plt.subplots(figsize = (5,5))
    ax.plot([-5,5],[-5,5], c = 'r', ls = '--', lw = 0.5)
    if step:
        ax.step(x_array, f_array)
    else:
        ax.plot(x_array, f_array)
        ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.axvline(0, c = 'k')
    ax.axhline(0, c = 'k')
    ax.set_xticks(np.arange(-5, 5+1))
    ax.set_yticks(np.arange(-5, 5+1))
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    plt.grid(True)
    ax.set_aspect('equal', adjustable='box')
    fig.savefig("Figures"  + os.sep + title +'.svg', format='svg')

# 幂函数，p = 2
x_array = np.linspace(-5,5,1001)
f_array = np.power(x_array, 2)
visualize_fx(x_array, f_array, '幂函数_p=2')
# 反正弦函数
x_array_ = np.copy(x_array)
x_array_[(x_array_ < -1) | (x_array_ > 1)] = np.nan
f_array = np.arcsin(x_array_)
visualize_fx(x_array_, f_array, '反正弦函数')
# 正切函数
f_array = np.tan(x_array)
f_array[:-1][np.diff(f_array) < 0] = np.nan
visualize_fx(x_array, f_array, '正切函数')
# 向下取整函数
f_array = np.floor(x_array)
visualize_fx(x_array, f_array, '向下取整函数', True)
# 对数函数
x_array_ = np.copy(x_array)
x_array_[x_array_<=0] = np.nan
f_array = np.log(x_array_)
visualize_fx(x_array_, f_array, '对数函数')
```

## 统计运算

函数有:
* max
* min
* sum
* average
* var # 默认分母为 n 的方差
* std # 标准差
* argmin  # 返回数组中最小值索引的函数
* argmax  # 返回数组中最大值索引的函数
* cov   # 协方差矩阵(记得要转置)
* corrcoef  # 相关系数矩阵(也要转置)
* numpy.mean()  # 计算均值
* numpy.median() # 中位数
* numpy.nanmean() 
* numpy.nanmedian() # 忽略 NAN

*注意,这些统计函数都可设置 `axis` 这个参数,添加其时,相等于在这个维度上去进行评判,得到的不是一个标量*

**此外,注意几个特别的函数,mean**

```python
A_2D = np.random.randint(0,10, size = (4,6))
A_2D
```

```python
A_2D.max()
```

```python
A_2D.max(axis = 0)  # 此时得到的是每一列的最大值,所组成的向量
# axis = 0,说明沿着第一个轴的方向进行操作，即沿着行的方向进行操作，即对每一列进行操作!!!!(这里有点反直觉)
```

```python
A_2D.max(axis = 1)
```

```python
A_2D.max(axis = 1, keepdims = True) # 添加了后面参数 keepdims,可以确保得到的是二维数组
```

```python
indexs = np.argmax(A_2D, axis = 1)
x_indexs = np.argmax(A_2D, axis = 0)
print(indexs)
```

```python
print(A_2D.sum())
print(A_2D.sum(axis=0))
print(A_2D.sum(axis=1))
```
